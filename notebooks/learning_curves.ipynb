# Learning Curves

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataframe
df = pd.read_csv("../src/data/wandb_val_perf.csv")  # replace with your file path

# Extract columns
steps = df["Step"]

# Mouse model (floral-yogurt-181)
mouse_mean = df["floral-yogurt-181 - val_perf"]
mouse_min = df["floral-yogurt-181 - val_perf__MIN"]
mouse_max = df["floral-yogurt-181 - val_perf__MAX"]

# Human model (skilled-surf-178)
human_mean = df["skilled-surf-178 - val_perf"]
human_min = df["skilled-surf-178 - val_perf__MIN"]
human_max = df["skilled-surf-178 - val_perf__MAX"]

# --- Overlay plot only ---
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(steps, human_mean, label="Human Model", color="red")
ax.fill_between(steps, human_min, human_max, color="red", alpha=0.2)

ax.plot(steps, mouse_mean, label="Mouse Model", color="blue")
ax.fill_between(steps, mouse_min, mouse_max, color="blue", alpha=0.2)

ax.set_title("Learning Curves (Human vs Mouse)")
ax.set_xlabel("Step")
ax.set_ylabel("Validation Performance")
ax.legend()
ax.grid(True, linestyle="--", alpha=0.6)

plt.tight_layout()
plt.show()
```
